```python
df.tail()
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

In the code example I provided, I used the dropna() method with the inplace=True parameter to remove all rows with null values from the pandas dataframe. This method removes all rows containing any null values by default, unless specified otherwise.

The reason for adopting this approach is that null values can lead to inaccurate analysis and modeling. For instance, null values in numeric columns can cause errors in statistical calculations such as mean and standard deviation. Null values in categorical columns can result in missing categories and bias the analysis. Therefore, it is often best to remove rows with null values, especially if the null values are only a small portion of the dataset, in order to ensure accurate analysis and modeling. However, it is important to carefully consider the amount and distribution of null values in the dataset before deciding to remove them, as removing too many rows may result in loss of important information and affect the representativeness of the dataset.
